```python
class Circle:
    # Class Object Attribute
    PI = np.pi
    
    # Constructor
    def __init__(self, radius, center, color):
        self.radius = radius
        self.center = center
        self.color = color
        self.x = center[0]
        self.y = center[1]

    # Method:
    def get_circumference(self):
        return 2 * self.PI * self.radius
    
    def get_area(self):
        return Circle.PI * self.radius * self.radius
    
    def draw(self):
        draw_circle = plt.Circle(self.center, self.radius, color=self.color)
        fig, ax = plt.subplots()
        ax.add_patch(draw_circle)
        plt.xlim(self.center[0] - self.radius * 1.2, self.center[0] + self.radius * 1.2)
        plt.ylim(self.center[1] - self.radius * 1.2, self.center[1] + self.radius * 1.2)
        ax.set_aspect(1)
        plt.show()
```

```python
c = Circle(10, (2, 3), 'red')
c.draw()
```

```python
import inspect                                                                                                                             

def froze_it(cls):                                                                                                                      

    def frozensetattr(self, key, value):                                                                                                   
        if not hasattr(self, key) and inspect.stack()[1][3] != "__init__":                                                                 
            print("Class {} is frozen. Cannot set {} = {}"                                                                                 
                  .format(cls.__name__, key, value))                                                                                       
        else:                                                                                                                              
            self.__dict__[key] = value                                                                                                     

    cls.__setattr__ = frozensetattr                                                                                                        
    return cls                                                                                                                             

@froze_it                                                                                                                                  
class A:                                                                                                                                   
    def __init__(self):                                                                                                                    
        self._a = 0                                                                                                                        

a = A()                                                                                                                                    
a._a = 1                                                                                                                                   
a._b = 2 # error
```

```python
# Base clas -> Animal
# Derived -> Dog
class Dog(Animal):
    def __init__(self):
        print("Dog Created")
        self.number_of_legs = 4
        Animal.__init__(self)
```

```python
class Line:
    def __init__(self, coor1, coor2):
        self.coor1 = coor1
        self.coor2 = coor2
        
    def distance(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        
        return ((x2- x1) ** 2 + (y2-y1) ** 2) ** .5
    
    def slope(self):
        x1, y1 = self.coor1
        x2, y2 = self.coor2
        return (y2-y1) / (x2-x1)
        
```

```python
coordinate1 = (0, 0)
coordinate2 = (10, 0)

li = Line(coordinate1, coordinate2)
```

```python
li.slope()
```

```python
li.distance()
```
