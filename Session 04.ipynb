```python
# Dehghan
# Roman to Numeric
roman_val = {
    "I": 1, 
    "V": 5, 
    "X": 10, 
    "L":50,
    "C": 100,
    "D": 500,
    "M": 1000
}
roman_str = input("Enter ROMAN number:")
integer_value = 0
for index in range(len(roman_str)):
    # 
    if index > 0 and roman_val[roman_str[index]] > roman_val[roman_str[index-1]]:
        integer_value = integer_value + roman_val[roman_str[index]] - 2* roman_val[roman_str[index-1]]
    else:   # 
        integer_value = integer_value + roman_val[roman_str[index]]
print(integer_value)
```

```python
# Esmaili
# Numeric to Roman
roman_chars_symbol = ["I","IV","V","IX","X","XL","L","XC","C","CD","D","CM","M"]
roman_chars_value = [1,4,5,9,10,40,50,90,100,400,500,900,1000]

number = int(input("please enter a number:"))
print(f"number '{number}' = ",end = "")

result = ""
while number > 0:
    for i in range(len(roman_chars_value)-1,-1,-1):
        if number >= roman_chars_value[i]:
            result += roman_chars_symbol[i]
            number -= roman_chars_value[i]
            
print(result)
```

```python
# Playing Cards:
SUITS = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
RANKS = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
```

```python
suit = random.randrange(0, len(SUITS))
rank = random.randrange(0, len(RANKS))
print(f'Your Card is {RANKS[rank]} of {SUITS[suit]}')
name = f'{RANKS[rank]}_of_{SUITS[suit]}'.lower()
plt.figure(figsize=(10, 10))
card = plt.imread(f"Data\Cards\{name}.png")
plt.imshow(card)
plt.axis('off')
plt.show()
```

```python
suit = random.choice(SUITS)
rank = random.choice(RANKS)
print(f'Your Card is {rank} of {suit}')
name = f'{rank}_of_{suit}'.lower()
plt.figure(figsize=(10, 10))
card = plt.imread(f"Data\Cards\{name}.png")
plt.imshow(card)
plt.axis('off')
plt.show()
```

```python
# Create a deck of cards:
deck = []
for suit in SUITS:
    for rank in RANKS:
        s = f"{rank}_of_{suit}"
        deck += [s]
    
deck
```

```python
fig, axs = plt.subplots(nrows=4, ncols=13, figsize=(20, 20))
axs = axs.flatten()
for c, ax in zip(deck, axs):
    card = plt.imread(f"Data\Cards\{c}.png")
    ax.imshow(card)
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```

```python
from random import randint

SUITS = ['Clubs', 'Diamonds', 'Hearts', 'Spades']
RANKS = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jacl', 'Queen', 'King', 'Ace']
collected_suits = []
collected_cards = []
counter = 0

deck = []
for suit in SUITS:
    for rank in RANKS:
        deck.append('%s %s'%(suit,rank)) 

while len(collected_suits) != 4 :
    rand = randint(0,51)
    card = deck[rand].split()
    suit_of_card = card[0]
    if card[0] not in collected_suits:
        collected_suits.append(card[0])
    collected_cards.append(deck[rand])
    counter += 1
print(counter,collected_cards)
```

```python
n = int(input("How Many Numbers?"))
count = 0
collected_count = 0
is_collected = [False] * n
while collected_count != n:
    new_card = random.randrange(0, n)
    count += 1
    if not is_collected[new_card]:
        is_collected[new_card] = True
        collected_count += 1
print(count)
```
