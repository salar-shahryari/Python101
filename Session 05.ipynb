```python
# Dehghani
# exersice no.1 - session.4
count_list = []
x = 100  # number of iterations
n = int(input("How Many Numbers?")) # number of SUITS!
for num in range(x):
    count = 0
    collected_count = 0
    is_collected = [False] * n
    while collected_count != n:
        new_card = random.randrange(0, n)
        count += 1
        if not is_collected[new_card]:
            is_collected[new_card] = True
            collected_count += 1
    count_list.append(count)
count_ave = round(sum(count_list) / x)
print (f"the average of counter for {x} try is {count_ave}")
```

```python
n = int(input("Grid Size?"))
a = [[False] * n for _ in range(n)]
dead_end = 0
x = n//2
y = n//2
path = [(x, y)]
while 0 < x < n -1 and 0 < y < n -1:
    a[x][y] = True
    if a[x-1][y] and a[x+1][y] and a[x][y-1] and a[x][y+1]:
        dead_end += 1
        break
    r = random.randint(1, 4)
    if r == 1 and not a[x+1][y]:
        x += 1
    elif r == 2 and not a[x-1][y]:
        x -= 1
    elif r == 3 and not a[x][y+1]:
        y += 1
    elif r == 4 and not a[x][y-1]:
        y -= 1
    if (x, y) not in path:
        path += [(x, y)]
        
print(path)
```

```python
plt.figure(figsize=(12, 12))
xticks = [i for i in range(n)]
yticks = [i for i in range(n)]
plt.plot()
plt.xticks(xticks)
plt.yticks(yticks)
plt.axis('equal')
for i in range(len(path) - 1):
    X = [path[i][0], path[i+1][0]]
    Y = [path[i][1], path[i+1][1]]
    plt.plot(X, Y, 'bo-', linewidth=1)
plt.xlim(-1, n)
plt.ylim(-1, n)
plt.plot(n//2, n//2, 'go', ms =12, label='start')
plt.plot(path[-1][0], path[-1][1], 'ro', ms =12, label='end')
plt.axhline(0, color='r', ls='--', alpha=.7)
plt.axhline(n-1, color='r', ls ='--', alpha=.7)
plt.axvline(0, color='r', ls ='--', alpha=.7)
plt.axvline(n-1, color='r', ls ='--', alpha=.7)
plt.legend(loc='upper right')
plt.grid()
plt.show()
```
