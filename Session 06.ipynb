```python
import matplotlib.pyplot as plt
import random
import math
```

```python
m = int(input('How many iterations?'))
n = int(input("Grid Size?"))
dead_end = 0
for i in range(m):
    a = [[False] * n for _ in range(n)]
    x = n//2
    y = n//2
    while 0 < x < n -1 and 0 < y < n -1:
        a[x][y] = True
        if a[x-1][y] and a[x+1][y] and a[x][y-1] and a[x][y+1]:
            dead_end += 1
            break
        r = random.randint(1, 4)
        if r == 1 and not a[x+1][y]:
            x += 1
        elif r == 2 and not a[x-1][y]:
            x -= 1
        elif r == 3 and not a[x][y+1]:
            y += 1
        elif r == 4 and not a[x][y-1]:
            y -= 1
print(f"Dead End Percentage: {dead_end / m * 100}%")
```

```python
def my_rand(my_low=0, my_high=10, nums=1):
    """determine the random numbers (nums) between my_low to my_high numbers
    my_low = the min value that can be produced (inclusive)
    my_high = the max value that can be produced (inclusive)
    """
    x = []
    for item in range(nums):
        x.append(random.randint(my_low, my_high))
    return x
```

```python
def rand_in_range(low, high):
    d = abs(low-high)
    minus = abs(low)
    return random.random() * d - minus
```

```python
def uniform_sample(my_low=0, my_high=1, nums=None):
    """determine the random numbers (nums) between my_low to my_high numbers"""
    if nums is None:
        return rand_in_range(my_low, my_high)

    else:
        x = []
        for _ in range(nums):
            x.append(rand_in_range(my_low, my_high))
        return x
```

```python
max(uniform_sample(my_low=-100, my_high=1, nums=100000))
```

```python
def randint_sample(my_low, my_high, nums=None):
    """
    my_high (exclusive)
    """
    mylist = uniform_sample(my_low, my_high, nums)
    return [math.floor(item) for item in mylist]
```

```python
def randint_sample2(my_low, my_high, nums=None):
    """
    my_high (inclusive)
    """
    mylist = uniform_sample(my_low, my_high+1, nums)
    return [math.floor(item) for item in mylist]
```
