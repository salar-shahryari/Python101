# map()

```python
# map(function, iterable, ...)
```

```python
def forenheit(celsius):
    return (9/5) * celsius + 32
```

```python
temp = (0, 22.5, 40.0, 100.0)
```

```python
f_temps = map(forenheit, temp)
```

```python
list(f_temps)
```

```python
list(map(lambda celsius:(9 / 5) * celsius + 32, (0, 22.5, 40.0, 100.0)))
```

```python
a = [1,2,3,4]
b = [5,6,7,8]
c = [10, 11, 12, 13]
```

```python
list(map(lambda x, y: x + y, a, b))
```

```python
list(map(lambda x, y, z: x + y + z, a, b, c))
```

# enumerate()

```python
# def enumerate(sequence, start=0):
#     n = start
#     for elem in sequence:
#         yield n, elem
#         n += 1
```

```python
lst = 'a b c'.split()
```

```python
for number, item in enumerate(lst):
    print(number, item)
```

```python
for number, item in enumerate(lst, start=10):
    print(number, item)
```

```python
lst1 = 'a b c'.split()
lst2 = ('tehran', 'tabriz', 'shiraz')
```

```python
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
```

```python
for _ in d1:
    print(_)
```

```python
list(zip(d1, d2))
```

```python
list(zip(d1.values(), d2))
```

```python
list(zip(d2, d1.values()))
```

```python
for _ in d1:
    print(_)
```

```python
for k in d1.keys():
    print(k)
```

```python
for v in d1.values():
    print(v)
```

```python
for k, v in d1.items():
    print(k, v)
```

```python
def switch(d1, d2):
    result = {}
    for d1key, d2val in zip(d1, d2.values()):
        result[d1key] = d2val
    return result   
```

```python
switch(d1, d2)
```
