```python
def random_walk(steps):
    x, y = 0, 0
    for _ in range(steps):
        dx, dy = random.choice([(0, 1), (0, -1), (1, 0), (-1, 0)])
        x += dx
        y += dy
    return x, y
```

```python
def manhathan_distance(x, y):
    return abs(x) + abs(y)
```

# What is the longest random walk you can take so that on average you will end up 4 blocks or fewer from home...

```python
number_of_walks = 100000
all_no_transport_percentages = np.zeros(31)
for walk_length in range(1, 31):
    no_transport = 0 # number of walks four or fewer from home
    for i in range(number_of_walks):
        x, y = random_walk(walk_length)
        distance = manhathan_distance(x, y)
        if distance <= 4:
            no_transport += 1
    no_transport_percentage = no_transport / number_of_walks
    all_no_transport_percentages[walk_length] = no_transport_percentage
    print(f"Walk Size: {walk_length:2}, percentage of no transport: {round(no_transport_percentage * 100, 3)}%")
```
