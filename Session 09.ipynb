```python
import math
import random

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Esmaili:
def f(v0, a, x0, y0, x):
    """
    get height
    :param v0: initial speed
    :param a: initial angle
    :param x0: initial x position
    :param y0: initial y position
    :param x: current location
    :return:
    """
    g = 9.8
    return (-g * (x - x0) ** 2) / (2 * v0 ** 2 * math.cos(a) ** 2) + (x - x0) * math.tan(a) + y0


v0 = float(input("V0 (initial speed) (m/s):"))
a = float(input("alpha (initial angle) (degree):"))
x0 = float(input("x0 (initial x position):"))
y0 = float(input("y0 (initial y position):"))

x = []
y = []
current_x = x0
current_y = y0
while current_y >= 0:
    x.append(current_x)
    y.append(current_y)

    current_x += 1
    current_y = f(v0, math.radians(a), x0, y0, current_x)

plt.plot(x, y)
plt.plot([x0], [y0], 'go--')
plt.plot([x[-1]], [y[-1]], 'ro--')
plt.grid()
plt.show()
```

```python
# Pur Bagher:
v_0 = int (input ('Please enter the initial velocity: '))
theta_0 = int (input ('Please enter the projectile angle wrt the horizontal axis: '))
y_0 = int (input ('Please enter the initial eight: '))
# Defining some constants (initial position and the gravitational force)
sx_0 = 0; g = 9.81
t = np.linspace(0, 30, 1000)
sy_d = v_0 * np.sin (np.deg2rad(theta_0)) * t - 0.5 * g * t ** 2 + y_0
sy_d = sy_d[sy_d>= 0]
t = t[:len(sy_d)]
plt.plot (t, sy_d, label = 'Y Position')
plt.title ('Projectile Motion as a Function of Time')
plt.legend ()
plt.grid ()
plt.show ()
```

```python
# Behruz:
#Initial throw speed
v0 = 2
#Initial height
h0 = 2
fall_time = (h0/9.8)**0.5
t = np.linspace(0, fall_time, 10)
def f(t):
    return v0 * t
def f2(t):
    return h0 - (9.8 * (t ** 2))
plt.plot(f(t), f2(t))
plt.grid()
plt.show()
print (f"It hits the ground in {round(fall_time*v0,2)} ")
```

```python
def z_function(x, y):
    return 50 - (x ** 2 + y ** 2)
```

```python
n = 10
x_values = np.linspace(-5, 5, n)
y_values = np.linspace(-5, 5, n)
```

```python
x_values
```

```python
X, Y = np.meshgrid(x_values, y_values)
```

```python
X
```

```python
plt.scatter(X, Y)
```

```python
Z = z_function(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
