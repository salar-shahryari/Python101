# Pandas Merge

```python
registration = pd.DataFrame({'reg_id': [1, 2, 3, 4], 'name': ['Andrew', 'Bob', 'Charlie', 'David']})
```

```python
registration
```

```python
logins = pd.DataFrame({'login_id': [1, 2, 3, 4], 'name': ['Xavier', 'Andrew', 'Yolanda', 'Bob']})
```

```python
logins
```

## Inner Join

```python
pd.merge(registration, logins, how='inner', on='name')
```

```python
# if only one column name matches
pd.merge(registration, logins, how='inner')
```

```python
pd.merge(registration, logins, how='inner', on='reg_id')
```

# Left, Right Join

```python
pd.merge(registration, logins, how='left', on='name')
```

```python
pd.merge(registration, logins, how='right', on='name')
```

# Outer Join

```python
pd.merge(registration, logins, how='outer', on='name')
```

```python
registration = registration.set_index('name')
```

```python
registration
```

```python
# left_index : bool, default False, Use the index from the left DataFrame as the join key(s) -> True
# right_on : label or list, or array-like Column or index level names to join on in the right DataFrame.
pd.merge(registration, logins, left_index=True, right_on='name')
```

```python
registration = registration.reset_index()
```

```python
registration
```

```python
registration.columns = ['reg_name', 'reg_id']
```

```python
registration
```

```python
pd.merge(registration, logins, how='inner', left_on='reg_name', right_on='name').drop('reg_name', axis=1)
```

```python
registration.columns = ['name', 'id']
registration
```

```python
logins
```

```python
logins.columns = ['id', 'name']
logins
```

```python
pd.merge(registration, logins, on='name')
```

```python
pd.merge(registration, logins, on='name', suffixes=('_reg', '_log'))
```

```python
messy_names = pd.Series(['andrew    ', 'b;;ob', '    charlie'])
```

```python
def cleanup(name):
    name = name.replace(';', '')
    name = name.strip()
    name = name.capitalize()
    return name
```

```python
messy_names.apply(cleanup)
```

```python
np.vectorize(cleanup)(messy_names)
```
