```python
# Binary to Decimal, Method 1:
n = int(input("Enter the number:"))
t = 0 # totall
i = 0
while n != 0:
    t += (n % 10) * 2 ** i
    n //= 10
    i += 1
print(t)
```

```python
# Decimal to Binary, Method 1:
n = int(input("Enter the number:"))
v = 1
while v <= n//2:
    v *= 2
while v > 0:
    if n < v:
        print(0, end="")
    else:
        print(1, end="")
        n -= v
    v //= 2
print()
```

```python
# Decimal to Binary, Method 2:
input_number=int(input("please enter a decimal number:"))
result=""

number = input_number
while number > 0:
    result = result + str(number % 2)
    number = number // 2

result=result[::-1]
print(F"{input_number} in decimal = {result} in binary")
```

```python
# Checkers Pattern:
# * * * * *
#  * * * *
# * * * * *
#  * * * *
# * * * * *
line = ''
for i in range (5):
    if i % 2 == 0 :
        for i2 in range(9):
            if i2 % 2 == 0:
                line += '*'
            else :
                line += ' '
    else :
        for i2 in range(9):
            if i2 % 2 == 1:
                line += '*'
            else :
                line += ' '
    line += '\n'
print(line)
```

```python
stake=int(input("please enter stake:"))
goal=int(input("please enter goal:"))
trials=int(input("please enter trials:"))

count_win = 0
count_loss = 0
bet = 0

for i in range(trials):
    current_stake = stake
    while 0 < current_stake < goal:
        bet += 1
        coin = random.randint(0,1)
        if coin == 0:
            current_stake -= 1
        else:
            current_stake += 1
    
    if current_stake == 0:
        count_loss += 1
    if current_stake == goal:
        count_win += 1
        
print(f"Result : win = {count_win / trials * 100 }% , loss = {count_loss / trials * 100}%")
print(f"Result : betting average: {bet/trials}")
```
